```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../DatosPrueba/Generacion_(kWh)_2023.csv' , header=2)
```

```python
df
```

```python
#Convertir el Campo fecha Operacion en Tipo Fecha
df['Fecha'] = pd.to_datetime(df['Fecha'])
```

```python
#Sacar el Mes de la fecha 
df['Mes'] = df['Fecha'].dt.month
```

```python
#Cantidad de generacion por mes
generacion_total_mes = df.groupby('Mes')['Combustible'].count()
```

```python
generacion_total_mes
```

```python
#cantidad de generacion ordenado por mes y por fuente
generacion_fuente_mes = df.groupby(['Mes', 'Combustible'])['Combustible'].count()
```

```python
generacion_fuente_mes
```

```python
#calculamos el porcentaje de participacion de cada fuente (mensual)
porcentaje_participacion = (generacion_fuente_mes / generacion_total_mes)*100
```

```python
porcentaje_participacion
```

```python
#Creacion de nueva tabla
new_table = porcentaje_participacion.unstack().fillna(0)
```

```python
new_table
```

```python
#Cambiar nombre a las columnas
new_table =new_table.rename(columns={'ACPM': 'Porcentaje ACPM', 'AGUA': 'Porcentaje Agua', 'BAGAZO': 'Porcentaje Bagazo',
                                        'BIOGAS': 'Porcentaje Biogas', 'CARBON': 'Porcentaje Carbon', 'COMBUSTOLEO': 'Porcentaje Combustoleo',
                                        'CRUDO': 'Porcentaje Crudo', 'GAS': 'Porcentaje Gas', 'GAS NI': 'Porcentaje Gas NI',
                                        'JET-A1': 'Porcentaje JET-A1', 'RAD SOLAR': 'Porcentaje RAD SOLAR', 'VIENTO': 'Porcentaje Viento'})
```

```python
new_table
```

```python
new_table.to_csv('pregunta4_tabla.csv')
```
