```python
import pandas as pd
```

```python
#Leyendo Archivo de Generacion 2022
df2022 = pd.read_csv('../DatosPrueba/Generacion_(kWh)_2022.csv' , header=2)
```

```python
#Leyendo Archivo de Generacion 2023
df2023 = pd.read_csv('../DatosPrueba/Generacion_(kWh)_2023.csv' , header=2)
```

```python
#Concatenamos los dos Datasets
dfTotal = pd.concat([df2022,df2023])
```

```python
dfTotal.reset_index(drop=True)
```

```python
#Generamos archivo con nuevo Dataset que contiene la informacion de los dos archivos anteriores
dfTotal.to_csv('DatosGeneracion_(kWh)_2022_2023.csv', header=True)
```
